```python
import numpy as np
import torch
import tensorflow as tf

# 1. Membuat vector 1x3
# NumPy
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

# PyTorch
x_pt = torch.tensor([1, 2, 3])
y_pt = torch.tensor([4, 5, 6])

# TensorFlow
x_tf = tf.constant([1, 2, 3])
y_tf = tf.constant([4, 5, 6])

```

```python
# 2. Menghitung Dot Product
dot_numpy = np.dot(x, y)
dot_pytorch = torch.dot(x_pt, y_pt)
dot_tensorflow = tf.reduce_sum(tf.multiply(x_tf, y_tf))

print(f"Dot Product NumPy: {dot_numpy}")
print(f"Dot Product PyTorch: {dot_pytorch.item()}")
print(f"Dot Product TensorFlow: {dot_tensorflow.numpy()}")
```

```python
import numpy as np

# 1. Transpose Matriks Y
# Mendefinisikan matriks Y
Y = np.array([[42, 4, 7, 99],
              [-99, -3, 17, 22]])

# Melakukan transpose
Y_transpose = Y.T

print("1. Transpose dari Matriks Y:")
print(Y_transpose)
print("-" * 30)


# 2. Hadamard Product
# Mendefinisikan dua matriks 2x2
A = np.array([[25, 10],
              [-2, 1]])

B = np.array([[-1, 7],
              [10, 8]])

hadamard_res = A * B

print("2. Hadamard Product dari Matriks:")
print(hadamard_res)
print("-" * 30)


# 3. Dot Product Tensor w dan x
# Mendefinisikan vektor w dan x
w = np.array([-1, 2, -2])
x = np.array([5, 10, 0])

# Menghitung dot product
dot_res = np.dot(w, x)

print("3. Dot Product dari Tensor w dan x:")
print(f"Hasil: {dot_res}")
```
