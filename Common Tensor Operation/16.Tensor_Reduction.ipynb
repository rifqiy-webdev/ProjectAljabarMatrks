```python
import numpy as np
import torch
import tensorflow as tf
```

```python
#buat matrix X dengan ukuran 2x3 menggunakan numpy
X = np.array([[25, 55, 61], [13, 1, 7]])
print("X=", X)
```

```python
#hitung berapa hasil penjumlahan elemen matrix X dengan fungsi sum
X.sum()
print("hasil penjumlahan elemen matrix X:", X.sum())
```

```python
#buat matrix X_tf dengan ukuran 2x3 menggunakan tensorflow
X_tf = tf.Variable([[25, 55, 61], [13, 1, 7]])
print("X=", X_tf)
```

```python
#buat matrix X_pt dengan ukuran 2x3 menggunakan pytorch
X_pt = torch.tensor([[25, 55, 61], [13, 1, 7]])
print("X_pt=", X_pt)
```

```python
#hitung hasil penjumlahan seluruh elemen matrix X, X_pt, X_tf
sum_X = np.sum(X)
sum_X_pt = torch.sum(X_pt)
sum_X_tf = tf.reduce_sum(X_tf)

print(f"Sum All - NumPy: {sum_X}, PyTorch: {sum_X_pt.item()}, TF: {sum_X_tf.numpy()}")
```

```python
#hitung hasil penjumlahan matrix X, X_pt, X_tf berdasarkan row nya
X.sum(axis=1)
print("hasil penjumlahan matrix X berdasarkan rownya:", X.sum(axis=1))
torch.sum(X_pt, 1)
print("hasil penjumlahan matrix X berdasarkan rownya:", torch.sum(X_pt, 1))
tf.reduce_sum(X_tf, axis=1)
print("hasil penjumlahan matrix X berdasarkan rownya:", tf.reduce_sum(X_tf, axis=1))
```

```python
#hitung hasil penjumlahan matrix X, X_pt,X_tf berdasarkan kolom nya
X.sum(axis=0)
print("hasil penjumlahan matrix X berdasarkan rownya:", X.sum(axis=0))
torch.sum(X_pt, 0)
print("hasil penjumlahan matrix X berdasarkan rownya:", torch.sum(X_pt, 0))
tf.reduce_sum(X_tf, axis=0)
print("hasil penjumlahan matrix X berdasarkan rownya:", tf.reduce_sum(X_tf, axis=0))
```

```python
#hitung nilai minimal dari matrix X, X_pt dan X_tf
X.min()
print("nilai minimal dari matrix X adalah:", X.min())
print("nilai minimal dari matrix X adalah:", torch.min(X_pt))
print("nilai minimal dari matrix X adalah:", tf.reduce_min(X_tf))
```

```python
#hitung nilai maksimal dari matrix X, X_pt dan X_tf
X.max()
print("nilai maksimal dari matrix X adalah:", X.max())
torch.max(X_pt)
print("nilai maksimal dari matrix X adalah:", torch.max(X_pt))
tf.reduce_max(X_tf)
print("nilai maksimal dari matrix X adalah:", tf.reduce_max(X_tf))
```

```python
#hitung nilai rata-rata dari matrix X, X_pt dan X_tf
np.mean(X)
print("nilai rata-rata dari matrix X adalah:", np.mean(X))
torch.mean(X_pt.float())
print("nilai rata-rata dari matrix X adalah:", torch.mean(X_pt.float()))
tf.reduce_mean(X_tf)
print("nilai rata-rata dari matrix X adalah:", tf.reduce_mean(X_tf))
```

```python
#hitung hasil perkalian seluruh elemen matrix X
X.prod()
print("hasil perkalian seluruh elemen matrix X adalah:", X.prod())
```

```python
#hitung hasil perkalian seluruh elemen matrix X_pt
torch.prod(X_pt)
print("hasil perkalian seluruh elemen matrix X adalah:",torch.prod(X_pt))
```

```python
#hitung hasil perkalian seluruh elemen matrix X_tf
tf.reduce_prod(X_tf)
print("hasil perkalian seluruh elemen matrix X adalah:", tf.reduce_prod(X_tf))
```
