```python
import numpy as np
import torch
import tensorflow as tf
```

```python
#buat matrix X dengan ukuran 2x3 menggunakan numpy
X = np.array([[18, 81], [8, 1], [1, 8]])
X
```

```python
#buat matrix X_pt dengan ukuran 2x3 menggunakan pytorch
X_pt = torch.tensor([[18, 81], [8, 1], [1, 8]])
X_pt
```

```python
#buat matrix X_tf dengan ukuran 2x3 menggunakan tensorflow
X_tf = tf.Variable([[18, 81], [8, 1], [1, 8]])
X_tf
```

```python
#lakukan perkalian matrix X, X_pt dengan suatu skalar
X_pt*2
```

```python
#lakukan perkalian matrix X, X_tf dengan suatu skalar
X_tf*2
```

```python
#akukan penjumlahan matrix X, X_pt dan X_tf dengan suatu skalar
X_pt+4
```

```python
#akukan penjumlahan matrix X, X_pt dan X_tf dengan suatu skalar
X_tf+4
```

```python
#akukan pengurangan matrix X, X_pt dan X_tf dengan suatu skalar
X_pt-6
```

```python
#akukan pengurangan matrix X, X_pt dan X_tf dengan suatu skalar
X_tf-6
```
