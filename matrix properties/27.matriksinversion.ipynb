```python
#buat matrix X ukuran 2x2 menggunakan numpy
X = np.array([[-1, 8], [18, -81]])
X

```

```python
#buat invers dari matrix X dan beri nama Xinv menggunakan numpy
Xinv = np.linalg.inv(X)
Xinv
```

```python
#kalikan X dengan Xinv menggunakan numpy
np.dot(Xinv, X)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import torch
import tensorflow as tf
```

```python
#Numpy
A = np.array([[4,2], [-5,-3]])
B = np.array([[4,-7]])
A
A_inv = np.linalg.inv(A)
C = A_inv.dot(B.T)
C
```

```python
#Pytorch
x = torch.tensor([[4.0, 2.0],[-5.0, -3.0]] )

y = torch.tensor([[4.0], [-7.0]])
x_inv = torch.linalg.inv(x)

solution = x_inv @ y

print(solution)
```
