```python
import numpy as np

# Membuat matriks 3x3
matrix_np = np.array([[10, 2, 3],
                      [4, 15, 6],
                      [7, 8, 9]])

# Menghitung Frobenius Norm
frobenius_np = np.linalg.norm(matrix_np, ord='fro')

print(f"NumPy Frobenius Norm: {frobenius_np}")
```

```python
import torch

# Membuat matriks 3x3 (Tensor)
matrix_pt = torch.tensor([[10., 2., 3.],
                          [4., 15., 6.],
                          [7., 8., 9.]])

# Menghitung Frobenius Norm
frobenius_pt = torch.norm(matrix_pt, p='fro')

print(f"PyTorch Frobenius Norm: {frobenius_pt.item()}")
```

```python
import tensorflow as tf

# Membuat matriks 3x3 (Constant)
matrix_tf = tf.constant([[10.0, 2.0, 3.0],
                         [4.0, 15.0, 6.0],
                         [7.0, 8.0, 9.0]])

# Menghitung Frobenius Norm
frobenius_tf = tf.norm(matrix_tf, ord='euclidean')

print(f"TensorFlow Frobenius Norm: {frobenius_tf.numpy()}")
```
