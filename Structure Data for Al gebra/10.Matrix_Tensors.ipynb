SOAL NOMOR 1

```python
# buat matrix X berukuran 2x3 menggunakan numpy
import numpy as np

x = np.array([[1, 2, 3], [3, 2, 1]])
print (x)

# tampilkan ukuran dari matrix X
print ("Ukuran matriks x :", x.shape)

# tampilkan seluruh elemen dari kolom pertama dari matrix X
print ("Elemen kolom pertama :", x[0,:])

# tampilkan seluruh elemen dari baris kedua dari matrix X
print ("Elemen kolom kedua :", x[1,:])
```

SOAL NOMOR 2

```python
# ulangi soal nomor 1 menggunakan pytorch
import torch

x_pt = torch.tensor([[1, 2, 3], [3, 2, 1]])
print (x_pt)

# tampilkan ukuran dari matrix x_pt
print ("Ukuran matriks x :", x_pt.shape)

# tampilkan seluruh elemen dari kolom pertama dari matrix x_pt
print ("Elemen kolom pertama :", x_pt[0,:])

# tampilkan seluruh elemen dari baris kedua dari matrix x_pt
print ("Elemen kolom kedua :", x_pt[1,:])
```

```python
# ulangi soal nomor 1 menggunakan tensorflow
import tensorflow as tf

x_tf = tf.Variable([[1, 2, 3], [3, 2, 1]])
print (x_tf)

# tampilkan ukuran dari matrix x_tf
tf.shape(x_tf)

# tampilkan seluruh elemen dari kolom pertama dari matrix x_pt
print ("Elemen kolom pertama :", x_tf[0,:])

# tampilkan seluruh elemen dari baris kedua dari matrix x_pt
print ("Elemen kolom kedua :", x_tf[1,:])
```
