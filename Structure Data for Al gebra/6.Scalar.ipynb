```python
import tensorflow as tf

# lakukan tahapan pada soal 1 dengan menggunakan tensorflow

# buat variabel skalar x dan simpan suatu nilai skalar
x_tf = tf.Variable(57, dtype = tf.int32)
x_tf
```

```python
# buat variabel skalar y dan simpan nilai skalar serta tipe datanya
y_tf = tf.Variable(75, dtype = tf.int32)
y_tf
```

```python
# jumlahkan x_tf + y_tf
tf_sum = tf.add(x_tf, y_tf)
tf_sum
```

```python
# buat variabel bernama x_float_tf dan simpan suatu nilai desimal serta tipe datanya
x_float_tf = tf.Variable(57.8, dtype = tf.float32)
x_float_tf
```

```python
import torch as pt

# lakukan tahapan pada soal 1 dengan menggunakan pytorch

# buat variabel skalar x dan simpan nilai skalar serta tipe datanya
x_pt = pt.tensor(57, dtype = pt.int32)
x_pt
```

```python
# buat variabel skalar y dan simpan nilai skalar serta tipe datanya
y_pt = pt.tensor(75, dtype = pt.int32)
y_pt
```

```python
# jumlahkan x_pt + y_pt
pt_sum = pt.add(x_pt, y_pt)
pt_sum
```

```python
# buat variabel bernama x_float_pt dan simpan suatu nilai desimal serta tipe datanya
x_float_pt = pt.tensor(57.8, dtype = pt.float32)
x_float_pt
```
