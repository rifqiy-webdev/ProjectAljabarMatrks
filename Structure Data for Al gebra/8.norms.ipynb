unit vector


```python
#buat vector v dengan elemen (3,4)
v = np.array([1,8])

#hitung nilai L2 norm dari vector v
norm = np.linalg.norm(v)

#tampilkan unit vector dari vector v
unit_v = v/ norm

print("vector aseli:", v)
print("panjang vector:", norm)
print("unit vector:", unit_v) #unit vector dari vector v
print("panjang unit vector:", np.linalg.norm(unit_v)) # panjang unit vector dari vector v
```
